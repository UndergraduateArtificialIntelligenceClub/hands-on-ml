```python
import matplotlib
import numpy as np
%matplotlib inline
```

Usually matplotlib will output graphs in a new module. The second line lets them run inline.

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1, 2, 3, 6, 8, 12])
plt.show()
```

Notice the points graphed are (0,1), (1,2), (2, 3), (3,6), (4,8), (5,12). We can also pass two arrays to the plot function to give (x,y) coordinates, although it's slightly counterintuitive.

```python
plt.plot([-1, 3, 5, 7, 9], [21, 42, 17, 29, 30])
plt.show()
```

We can also adjust the axis markers by calling the .axis method with a list of [xmin, xmax, ymin, ymax], and adjusting xticks.

```python
plt.plot([-1, 3, 5, 7, 9], [21, 42, 17, 29, 30])
plt.axis([-2, 10, 15, 45])
plt.xticks(np.arange(-1, 10, 1.0))
plt.show()
```

We are now ready to define a function! To do this, we use numpy's linspace function, which will return the specified number of points, evenly distributed within that range. Notice this is similar to the arange method above, however, specifies the number of samples, versus step size.

```python
x=np.linspace(-2, 2, 400)
y=x**2
plt.plot(x,y)
plt.show()
```

To make this graph specier, we specify a title, labels, and add grid lines. 

```python
plt.plot(x,y)
plt.title("Quadratic Function")
plt.xlabel("x")
plt.grid(True)
plt.show()
```
